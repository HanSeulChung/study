# Question 3월달의 각 시간대별 온도의 평균들 중 가장 낮은 시간대의 온도를 출력하라

# Question 3월달의 각 시간대별 온도의 평균들 중 가장 높은 시간대의 온도를 출력하라

```python
#내가 생각한 답 근데 이게 아니라 시간대별로 높은온도 및 낮은 온도를 출력해야할것 같다.
target = df[df.Month==3].reset_index(drop=True)
answer_high = target.Temperature.max()
answer_low = target.Temperature.min()
print(f'3월달의 가장 높은 온도 :{answer_high}도')
print(f'3월달의 가장 낮은 온도 :{answer_low}도')
```

```python
# 그래서 다시 해보기
target = df[df.Month==3].reset_index(drop=True)
target['Hour'] = target.DateTime.dt.hour
target.head(3)
```

```python
target_m= target.groupby('Hour').Temperature.mean().to_frame().reset_index().sort_values('Temperature')
answer_low = target_m.head(1)
answer_high = target_m.tail(1)
print(answer_low)
print(answer_high)
```

```python
#답
target = df[df.DateTime.dt.month ==3]
result_low = target.groupby(target.DateTime.dt.hour)['Temperature'].mean().sort_values().values[0]
result_high = target.groupby(target.DateTime.dt.hour)['Temperature'].mean().sort_values().values[-1]
print(result_low)
print(result_high)
```

```python
df2 =df.copy()
def mkg(x):
    answer =""
    if x//10 ==0: 
        answer = "A"
    elif x//10 ==1:
        answer = "B"
    elif x//10 ==2:
        answer = "C"
    else:
        answer ="D"
    return answer
df2['grade'] = df2['Temperature'].apply(mkg)
answer = df2['grade'].value_counts()
answer 
```

```python
# 답
def split_data(x):
    if x<10:
        return "A"
    elif x<20:
        return 'B'
    elif x<30:
        return 'C'
    else:
        return 'D'
    
df['sp'] = df.Temperature.map(split_data)
result = df['sp'].value_counts()
display(result)
```

# Question Temperature의 평균이상의 Temperature의 값을 가지는 데이터를 Temperature를 기준으로 정렬 했을때 4번째 행의 Humidity 값은?

```python
Tm = df.Temperature.mean()
answer = df[df.Temperature >= Tm].sort_values('Temperature').reset_index(drop=True).Humidity[3]
answer
```

```python
# 답
result = df[df.Temperature >= df.Temperature.mean()].sort_values('Temperature').Humidity.values[3]
print(result)
```

# Question Temperature의 중간값 이상의 Temperature의 값을 가지는 데이터를Temperature를 기준으로 정렬 했을때 4번째 행의 Humidity 값은?

```python
Tmedian= df.Temperature.median()
answer = df[df.Temperature >= Tmedian].sort_values('Temperature').reset_index(drop=True).Humidity[3]
answer 
```

```python
# 답
result = df[df.Temperature >= df.Temperature.median()].sort_values('Temperature').Humidity.values[3]
print(result)
```
