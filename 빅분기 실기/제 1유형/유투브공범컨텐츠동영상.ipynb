# Question9 **2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오**

```python
video_df.ct =pd.to_datetime(video_df.ct)
target = video_df[(video_df.ct >="2021-11-01 00:00:00") & (video_df.ct<="2021-11-01 15:00:00")][['videoname','ct','viewcnt']]
target
```

```python
target2 = pd.concat([
            target.drop_duplicates('videoname',keep='first'),
            target.drop_duplicates('videoname',keep='last')])
target2=target2.sort_values('videoname').reset_index(drop=True)
target2['increase'] = target2.groupby('videoname').viewcnt.diff()
answer=target2.dropna()[['videoname','increase']]
answer
```

```python
video_df.ct =pd.to_datetime(video_df.ct)
target = video_df[(video_df.ct >="2021-11-01 00:00:00") & (video_df.ct<="2021-11-01 15:00:00")][['videoname','ct','viewcnt']]
target2 = pd.concat([
            target.drop_duplicates('videoname',keep='first'),
            target.drop_duplicates('videoname',keep='last')])
target2=target2.sort_values('videoname').reset_index(drop=True)
target2['increase'] = target2.groupby('videoname').viewcnt.diff()
answer=target2.dropna()[['videoname','increase']]
answer
```

```python
# 답
start = pd.to_datetime("2021-11-01 00:00:00")
end = pd.to_datetime("2021-11-01 15:00:00")

target = video_df.loc[(video_df["ct"] >= start) & (video_df['ct'] <= end)].reset_index(drop=True)

def check(x):
    result = max(x) - min(x)
    return result

answer = target[['videoname','viewcnt']].groupby("videoname").agg(check)
answer
```
