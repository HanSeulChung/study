```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
pd.set_option('max_columns',None)
pd.set_option('max_rows',None)
```

```python
X_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_test.csv')
```

```python
display(X_train.head())
display(y_train.head())
display(x_test.head())
```

```python
submission = pd.DataFrame(x_test.StudentID) #제출용 dataframe 생성
display(submission.head())
```

```python
df= pd.merge(X_train,y_train, on='StudentID')
X = df.drop(['StudentID','G3'],axis=1)
y = df.G3
x_test = x_test.drop('StudentID',axis=1)
display(X.head())
display(y.head())
```

```python
print(X.school.value_counts())
print(X.address.value_counts())
```

```python
print(X.info())
```

```python
# print(X.isnull().sum())
# print(y.isnull().sum())
X_obj = X.select_dtypes(include='object') # 거의 2개고 최대가 다섯개 이하이므로 labelencoding 진행
for col in X_obj.columns:
    print(X_obj[col].value_counts())
    
```

## LabelEncoding 진행

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

def LabelEncoding(df):
    obj = df.select_dtypes(include='object')
    for col in obj:
        df[col] = le.fit_transform(df[[col]])
    return df
X = LabelEncoding(X)
x_test = LabelEncoding(x_test)
display(X.head()) #no:0 yes:1로 확인 다른 string은 확인해야함
display(x_test.head())
```

## MinMax Scaling 진행

```python
from sklearn.preprocessing import MinMaxScaler
scaler_m = MinMaxScaler()

X_scaled= scaler_m.fit_transform(X)
x_test_scaled = scaler_m.fit_transform(x_test)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(X_scaled,y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()

model.fit(x_tr,y_tr)
pred_val=model.predict(x_val)
```

```python
from sklearn.metrics import r2_score

r2_score(y_val,pred_val)
```

```python
pred = model.predict(x_test_scaled)
```

```python
submission['pred'] = pred
submission['pred'] = submission['pred'].astype(int)
submission.to_csv('004001677.csv',index=False)
check = pd.read_csv('004001677.csv')
print(check.head())
```
